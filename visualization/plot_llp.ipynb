```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV data
file_path = "/root/CineSpark-Insights/results/rf_model_performance.csv"
df = pd.read_csv(file_path)

# Set the style of seaborn
sns.set(style="whitegrid")

# Plotting RMSE for different MaxDepth values
plt.figure(figsize=(12, 6))
sns.lineplot(x='MaxDepth', y='RMSE', hue='NumTrees', data=df, marker='o')
plt.title('RMSE vs MaxDepth for different NumTrees')
plt.show()

# Plotting MAE for different MaxBins values
plt.figure(figsize=(12, 6))
sns.lineplot(x='MaxBins', y='MAE', hue='NumTrees', data=df, marker='o')
plt.title('MAE vs MaxBins for different NumTrees')
plt.show()

# Plotting ExecutionTime for different NumTrees values
plt.figure(figsize=(12, 6))
sns.barplot(x='NumTrees', y='ExecutionTime', data=df, ci=None)
plt.title('ExecutionTime for different NumTrees')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV data
file_path = "/root/CineSpark-Insights/results/als_model_performance.csv"
df = pd.read_csv(file_path)

# Set the style of seaborn
sns.set(style="whitegrid")

# Plotting RMSE for different regParam values
plt.figure(figsize=(12, 6))
sns.lineplot(x='regParam', y='RMSE', hue='rank', data=df, marker='o')
plt.title('RMSE vs regParam for different rank')
plt.show()

# Plotting MAE for different maxIter values
plt.figure(figsize=(12, 6))
sns.lineplot(x='maxIter', y='MAE', hue='rank', data=df, marker='o')
plt.title('MAE vs maxIter for different rank')
plt.show()

# Plotting Training Time for different rank values
plt.figure(figsize=(12, 6))
sns.barplot(x='rank', y='Training Time', data=df, ci=None)
plt.title('Training Time for different rank')
plt.show()
```
